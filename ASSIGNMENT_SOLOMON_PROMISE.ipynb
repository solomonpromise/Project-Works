### Subsetting Iris_virginicas 

```python
#subsetting all species = Iris-virginica
iris_virginicas = iris[iris["species"] == "Iris-virginica"]
iris_virginicas.head()
```

```python
#subsetting Iris-virginica for sepal width >5 and >3
iris_vir5 = iris_virginicas[(iris_virginicas["sepal_length"] > 5.0) & (iris_virginicas["sepal_width"] > 3.0)]
iris_vir5
```

```python
#exporting the above subset
#I don't want my index, hence index = False
iris_vir5.to_csv("Iris_virg.csv", index = False, encoding = "utf8")
```

#### Subsetting Setosas

```python
#subsetting species = Iris-setosa
iris_setosa = iris[iris['species'] == "Iris-setosa"]
iris_setosa.head()
```

```python
#subsetting the species for petal length >= 3.50, & < 3.50
Iris_S = iris_setosa[(iris_setosa["petal_length"] >= 3.50) & (iris["petal_width"] < 3.50)]
Iris_S
```

### Reason for the empty table


**The subsetting the species for petal_length >= 3.50, & petal_width < 3.50 returned an empty table because for the first condtions (iris_setosa["petal_length"] >= 3.50) where all "FALSE", while that of the second conditions are "TRUE".
Recall that from the knowledge of boolean operation, TRUE AND FALSE equal FALSE, hence the reason for the empty table.**
