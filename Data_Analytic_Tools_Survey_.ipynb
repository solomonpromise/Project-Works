```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Rename columns
df.columns = ['Timestamp','Organisation', 'Role', 'Location', 'Visualization Tool', 'Database_Management Tool',
                  'Machine_Learning Tool', 'Big_Data Analytics Tool', 'Cloud Platform', 'Salary Range']

df.head()
```

```python
df.drop("Timestamp", axis=1, inplace=True)
df.head()
```

```python
tool_cols = [col for col in df.columns if 'Tool' in col]
salary_col = ['Salary Range']
cloud_col = [col for col in df.columns if col not in tool_cols+salary_col+['Organisation', 'Role', 'Location']]
other_col = ['Role', 'Location']
```

```python
# For cat cols and date cols fill in with mode and for num cols fill in with 9999
for col in df.columns:
  if col in df:
    df[col] = df[col].fillna(df[col].fillna('None'))
  elif col in salary_cols:
    df[col] = df[col].fillna(df[col].fillna('Prefer not to say'))
  elif col in cloud_col:
    df[col] = df[col].fillna(df[col].fillna('None'))

# Confirm that there aren't any missing values
df.isna().sum()
```
