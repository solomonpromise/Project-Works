# World Happiness Analysis

This work seeks to is to measure global happiness using the following variables: `life expectancy`, `economics`, `social support`, `absence of corruption`, `freedom`, and `generosity`.

The datasets used for this work is the World Happiness Report for the years 2015, 2016, 2017, 2018, and 2019, with which we seek to answer the following questions:

1. Which country is the happiest in the world?
2. What are the most important contributing factors to a nationâ€™s happiness?
3. Is overall happiness increasing or decreasing?

### Question 1: Which country is the happiest in the world?

**In an attempt to answer this question, several approaches/steps were carried out.**

**1.**: The combined dataset `df` was groupby `Year`, and the result of this is that the data was sorted in descending order for one year, follwed by the other. This result was stored in the variable name called `hap`.

```python
hap = df.groupby('Year')
hap.head(5)
```

**2:** The head which shows the first five happiest countires for each was aftered stored into a varible name called `hap_5`. This was to allow me carry out further analysis.

**Hecne the dataframe `hap_5` contains just the first five happiest countries for each year.**

```python
hap_5 = hap.head()
hap_5
```

```python
hap_5["Country"].unique()
```

**3:** The `hap_5`, was sorted to find the happiest country out of the five happiest countires for each year.

**After this sorting, `FINLAND` happens to be the happiest country with a score of `7.769` for the year 2019** 

```python
hap_5.sort_values("Score" , ascending=False)
```

**4:** Further Analysis was made based on the cummulative Score for the five years.

**After this, `DENMARK` seems to be the happiest conntry in the world with a cummulative score of `37.730` for the five years**

```python
hap_cum = hap_5.groupby('Country')['Score'].sum()
hap_cum
```

**5:** A dataframe was created to allow me visualize the countires with the happiest countires in the world based on their cummulative score for the five year 

```python
hap_cum = {'Country': ['Canada', 'Denmark', 'Finland', 'Iceland', 'Netherlands','Norway','Switzerland'], 'Happiness Score': [7.427, 37.730, 30.283, 37.555,7.488,37.705,30.077]}
hap_cum = pd.DataFrame(data=hap_cum)

hap_cum
```

### Visualization of top seven Happiest Countries in the World Based on their cummulative Score

```python
fig_dims = (10, 6)

fig, ax = plt.subplots(figsize=fig_dims)
sns.barplot(x = 'Country', y = 'Happiness Score', data=hap_cum)
plt.title("Top Seven Happiest Countiries in the world")
```

**Hence, I can say the regardless of year, `FINLAND` is the HAPPIEST COUNTIRES IN THE WORLD  with a score of `7.769`.**

**But if the cumulative score of the five year is taken as the basis of consideration, then `DENMARK` is the HAPPIEST COUNTRY IN THE WORLD WITH A TOTAL SCORE OF `37.730`.**

### QUESTION 2: What are the most important contributing factors to a nationâ€™s happiness?

A correlation heat map can be used to answer this question becuase it shows how related the variables within a dataset is. 

From the correlation heat map we can tell which variables are strongly related to the Happiness Score based on there correlation coefficient.

```python
X = df[[ 'GDP', 'Health', 'Support','Freedom','Generosity','Corruption']]
y = df['Score']
```

```python
corrMatrix = df.corr()
df.corr()
```

```python

fig_dims = (10, 6)
fig, ax = plt.subplots(figsize=fig_dims)
sns.heatmap(corrMatrix, annot=True)
plt.show()
```

From the heatmap, we four variables has a strong correlation with the happiness score, they include:
1. `GDP` with a coefficient of `0.75`
2. `Health` with a coefficient of `0.74`
3. `Support` with a coefficient of `0.7`
4. `Freedom` with a coefficient of `0.55`

There coefficient shows the degree to which they contribute to the happiness of a country, which `GDP` being the highest contributor, followed by `Health` etc.. 

**Secondly, a multiple linear regression model would help us quantify the contribution of each factor on the happiness of a country**

```python
# MULTIPLE LR
X = df[['GDP', 'Health', 'Support','Freedom','Generosity','Corruption']]
y = df['Score']

import statsmodels.api as sm
X_sm = X = sm.add_constant(X)
model = sm.OLS(y,X_sm)
model.fit().summary()
```

The OLS Summary above reveals that all the features are significant positive predictors or contributors to nations'happiness, even at 99% confidence level, except for `Generosity` with a coefficient of `0.2281`. The most contributing factors or features are `Freedom(1.3574)`, `Corruption(1.2839)`, `Health	(1.0597)`, `GDP(0.9681)`, and `Support(0.9172)`.

The result above means that a unit increase or change in people's freedom will result in an equaalent  increase in happiness by 1.3574 units. This applies to other too.
