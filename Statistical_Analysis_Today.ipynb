### Answer to Question four (4)

H0: The converted status is not dependent on the preferred language
H1: The converted status is dependent on the preferred language
    
Since both variables are categorical, Chi sqaure test of independency would be used to confirm if the converted status is not dependent or vice versa, using a significance level of 0.05

at the end of the test:
H0 would be accepted if the pvalue is equal to or greater than 0.05,
H1 would be accpeted if the p-value is less than 0.05.


At the end of the test, the p-value = 0.213 , since the p-value is greater than 0.05, we accept the H0 and reject the H1.
There we can conluded statistically that the CONVERTED STATUS IS NOT DEPENDENT ON THE PREFERRED LANGUAGE.

```python
cont_tab = pd.crosstab(df["converted"], df["language_preferred"])
cont_tab
```

```python
pd.crosstab(df["converted"], df["language_preferred"]).sum()
```

```python
pd.crosstab(df["converted"], df["language_preferred"]).plot(kind='bar')
```

```python
from scipy.stats import chi2_contingency
from scipy.stats import chi2
```

```python
cont_tab
```

```python
stat, p, dof, expected = chi2_contingency(cont_tab)
print('dof=%d' % dof ,"\n")
print("expected")
print(expected)


prob = 0.95
```

```python
# interpret p-value
alpha = 1.0 - prob
print('significance=%.3f, p=%.3f' % (alpha, p),"\n")
if p <= alpha:
	print('Dependent (reject H0)')
else:
	print('Independent (fail to reject H0)')
```

### Answer to Question 5.

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
dff = df[df['landing_page'] == 'new']
```

```python
mymod = ols('time_spent_on_the_page ~ language_preferred', data = dff).fit()  

# performing type 2 anova test  
aovtable = sm.stats.anova_lm(mymod, typ = 2)  
print('ANOVA table for New')  
print('----------------------')  
print(aovtable)  
print()  
```

H0: The is no scientific significant difference between the mean time spent on the new landing page 
    for the different preferred language
    
H1: The is a scientific significant difference between the mean time spent on the new landing page 
    for the different preferred language
    
Since there are three group in the categorical variable (language_preferred), one way anova would be 
used to confirm if there is a scientific significant difference between the mean time spent on the new landing page 
    for the different preferred language or vice versa.

at the end of the test:
H0 would be accepted if the pvalue is equal to or greater than 0.05,
H1 would be accpeted if the p-value is less than 0.05.


At the end of the test, the p-value = 0.432041 , since the p-value is greater than 0.05, we accept the H0 and reject the H1.
Therefore we can conluded that there is no scientific significant difference between the mean time spent on the new landing page for the different preferred language.
